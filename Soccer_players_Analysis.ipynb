```python
# loading the dataset and importing the necessary library

import pandas as pd

load = pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/fundamentals/football_players.csv', index_col='Name')
```

```python
tt = pd.DataFrame(load)
tt
```

```python
# looking at the attributes that shall be used to analyse the players

tt.columns
```

```python
# Getting to know the oldest active players

cc = ['Nationality', 'Age', 'Overall', 'Preferred Positions']
xt = tt[cc].sort_values('Age', ascending=False).head(15)
xt

# seeing that most of them are goalkeepers
```

```python
# getting to know the youngest active players

tx = ['Nationality', 'Age', 'Overall', 'Preferred Positions']
dd = tt[tx].sort_values('Age', ascending=True)[:15]
dd
```

```python
# to see how many young players are in each position

aa = dd['Preferred Positions'].value_counts()
aa = pd.DataFrame(aa)
aa


# most of them are aspiring to be offensive players
```

```python
# to know the top 15 countries in the world with the hghest number of footballers

pp = tt['Nationality'].value_counts()[:15]
pp =pd.DataFrame(pp)
pp
```

```python
# let's get to know the top 20 best players in the world

gg = ['Nationality', 'Age', 'Overall', 'Preferred Positions']

xx = tt[gg].sort_values('Overall', ascending=False)[:20]
xx
```

```python
# to determine who the strongest football players are

rr = ['Strength', 'Stamina', 'Aggression', 'Agility', 'Preferred Positions']

ee = tt[rr].sort_values('Strength', ascending = False)[:20]
ee = pd.DataFrame(ee)
ee


# seeing that most of them are defenders.
```

```python
# to determine the top 20 fastest players

ff = ['Dribbling', 'Reactions', 'Sprint speed', 'Acceleration', 'Ball control', 'Preferred Positions']

zz = tt[ff].sort_values('Sprint speed', ascending = False)[:20]
zz = pd.DataFrame(zz)
zz

# seeing that most of the fast players are versatile when it comes to positions
```

```python
# to determine the top 5 preferred positions

bn = tt['Preferred Positions'].value_counts()[:5]
bn = pd.DataFrame(bn)
bn

# showing us that most of the players are centre backs or famously known as defenders, then followed by goalkeepers
```

```python
# to see how many players are still active at the age of 40 and above

gf = tt[tt['Age'] >= 40].sort_values('Age', ascending=False)
gf


# there are only 16 players with most of them being GKs
```

```python
# to determine the top 20 best goalkeepers

gk = ['Nationality', 'Overall', 'GK diving','GK handling', 'GK kicking', 'GK positioning', 'GK reflexes', 'Preferred Positions']
gk = tt[gk].sort_values('Overall', ascending = False)
gk = gk[gk['Preferred Positions'].str.contains('GK')][:20]

gk
```
